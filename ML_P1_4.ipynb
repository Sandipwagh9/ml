```python
import pandas as pd
```

```python
df1 =pd.read_csv("/content/bank_marketing.csv")
df1.head()
```

```python
df1.drop(columns=["Unnamed: 0"],inplace=True,axis=1)
df1.head()
```

```python
df1.dropna()
```

One Hot Encoding

```python
encoded_df=pd.get_dummies(df1,columns=["job","marital","education","default","housing","loan"],drop_first=True)
```

```python
encoded_df.head()
```

```python
X=encoded_df.drop(columns=["contact","month","poutcome","deposit"],axis=1)
X.head()
```

```python
y=df1["deposit"]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=7)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

How many samples are misclassified by the model?

```python
misclassified_sample=(y_pred!=y_test).sum()
print(misclassified_sample)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
import pandas as pd
df2=pd.read_csv("/content/uber.csv")
```

```python
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2.dropna()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(data = df2[["fare_amount", "passenger_count"]])
plt.show()
```

```python
#removing outliers
q1=df2["fare_amount"].quantile(0.25)
q3=df2["fare_amount"].quantile(0.75)
iqr=q3-q1
lower=q1-1.5*iqr
upper=q3+1.5*iqr

df_new=df2[(df2["fare_amount"]>=lower) & (df2["fare_amount"]<=upper)]
df_new.head()
```

```python
df_new.dropna(axis=0)
X=df_new.drop(columns=["key","pickup_datetime","Unnamed: 0","fare_amount"],axis=1)
```

```python
y=df_new[["fare_amount"]]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python

model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model2=RandomForestRegressor(n_estimators=100,random_state=42)
model2.fit(X_train,y_train)
```

```python
r_y_pred=model2.predict(X_test)
```

```python
#Linear Regression
import numpy as np
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
print(mse)
rmse=np.sqrt(mse)
print(rmse)
```

```python
#RandomForest
mse=mean_squared_error(y_test,r_y_pred)
print(mse)
rmse=np.sqrt(mse)
print(rmse)
```
